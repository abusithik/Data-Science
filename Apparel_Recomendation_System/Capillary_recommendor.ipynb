```python
img_width, img_height = 128,128

top_model_weights_path = 'bottleneck_fc_model.h5' 
train_data_dir = 'C:/Users/smoham72/Desktop/Capillary/trainCapillary4/' 
nb_train_samples = 3026 
epochs = 50 
batch_size = 1
user_data=pd.read_csv('C:/Users/smoham72/Desktop/Capillary/trainCapillary4/train.csv')
```

```python
def save_bottlebeck_features():
    prods = []
    datagen = ImageDataGenerator(rescale=1. / 255)
    
    # build the VGG16 network
    model = applications.VGG16(include_top=False, weights='imagenet')
    generator = datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)
    
    for i in generator.filenames:
        prods.append(i[7:-4])

    bottleneck_features_train = model.predict_generator(generator, nb_train_samples // batch_size)
    bottleneck_features_train = bottleneck_features_train.reshape((3026,8192))
    np.save(open('C:/Users/smoham72/Desktop/Capillary/16k_data_cnn_features.npy', 'wb'), bottleneck_features_train)
    np.save(open('C:/Users/smoham72/Desktop/Capillary/16k_data_cnn_feature_asins.npy', 'wb'), np.asarray(prods))
save_bottlebeck_features()
```
